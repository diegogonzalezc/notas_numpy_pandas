# **Merge y concat**

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})

df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)# con el ignore_index se reinicia el indice para que no haya repetidos
```

```python
pd.concat([df1,df2],axis=1) # en caso que se quiera hacer por las columnas
```

```python
izq=pd.DataFrame({'key':['k4','k5','k6','k7'],
  'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7']})

der=pd.DataFrame({'key':['k4','k5','k6','k7'],
  'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
izq.merge(der, on = 'key') # aca estamos haciendo el merge en la columna key que es identica en los dos dataframes
```

```python
izq=pd.DataFrame({'key':['k4','k5','k6','k7'],
  'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7']})
der=pd.DataFrame({'key_2':['k4','k5','k6','k7'],
  'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
izq.merge(der, left_on='key',right_on='key_2')# em este caso le damos la instruccion de por donde se debe adherir en cada uno de los dataframes
```

```python
izq=pd.DataFrame({'key':['k4','k5','k6','k7'],
  'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7']})
der=pd.DataFrame({'key_2':['k4','k5','k6',np.nan],
  'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
izq.merge(der, left_on='key',right_on='key_2')# acá se está  haciendo un inner, por esta razon ya que key_2 no tiene k7 lo elimina
```

```python
izq.merge(der, left_on='key',right_on='key_2',how='right')# cuando le damos right le decimos que tome los datos de el dataframe der el cual no tiene k7
```

```python
izq=pd.DataFrame({'A':['A0','A1','A2'],
 'B':['B0','B1','B2']},
 index=['k0','k1','k2'])

der=pd.DataFrame({'C':['C0','C1','C2'],
 'D':['D0','D1','D2']},
 index=['k0','k2','k3'])
```

```python
izq.join(der)# este funciona por los indices, esa es la diferencia con merge y join
```

```python
izq.join(der, how='inner')# este funciona por los indices, esa es la diferencia con merge y join
```
